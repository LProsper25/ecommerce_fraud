### Featrures engineering

```python
df.loc[df["Customer Age"] < 0, "Customer Age"] = np.nan
df.dropna(axis=0, inplace=True)

df["High Amount"] = (df["Transaction Amount"] >= 800).astype(int)
df["Is Night"] = (df["Transaction Hour"] <= 5).astype(int)
df["New Account"] = (df["Account Age Days"] <= 30).astype(int)
df["Mismatch Address"] = (df["Shipping Address"] != df["Billing Address"]).astype(int)

cols_to_drop = ["Transaction ID", "Customer ID", "Product Category", "Transaction Date", "Customer Location", "IP Address", "Shipping Address", "Billing Address"]
df = df.drop(columns=cols_to_drop)
```

```python
df.head(10)
```

```python
# SÃ©paration X/y
y = df["Is Fraudulent"]
X = df.drop(columns="Is Fraudulent", axis=1)

# Train / Test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)
```

```python
# Preprocessing
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

# Selection of types
cat_features = X.select_dtypes(include=["object", "category"]).columns
num_features = X.select_dtypes(include=["float", "int"]).columns

# Encodage
cat_pipeline = Pipeline(
    [
        ("imputer", SimpleImputer(strategy='most_frequent')),
        ("encoder", OneHotEncoder(handle_unknown='ignore'))
    ]
)

# Scaling
num_pipeline = Pipeline(
    [
        ("imputer", SimpleImputer(strategy='median')),
        ("scaler", StandardScaler())
    ]
)

preprocessor = ColumnTransformer(
    [
        ("cat", cat_pipeline, cat_features),
        ("num", num_pipeline, num_features)
    ]
)
```

```python
# Pipeline
from sklearn.linear_model import LogisticRegression
modele = Pipeline(
    [
        ("preprocessing", preprocessor),
        ("model", LogisticRegression(class_weight='balanced'))
    ]
)
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay

modele.fit(X_train, y_train)
y_pred = modele.predict(X_test)
print(classification_report(y_test, y_pred))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
from sklearn.model_selection import GridSearchCV

# Use the Pipeline as the estimator (includes preprocessing)
modele_pipeline = Pipeline(
    [
        ("preprocessing", preprocessor),
        ("model", LogisticRegression())
    ]
)

grid_params = {
    'model__max_iter': [300, 500],
    'model__C': [0.01, 0.1, 1, 10],
    'model__class_weight': [None, 'balanced'],
    'model__random_state': [0, 40],
}

grid_search = GridSearchCV(modele_pipeline, grid_params, cv=5, scoring='f1', n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
y_proba = modele.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import precision_recall_curve
precision, recall, thresholds = precision_recall_curve(y_test, y_proba)
f1 = 2*(precision*recall)/(precision+recall)
best_threshold = thresholds[np.argmax(f1)]
best_threshold, f1.max()
```

```python
# Choix du seuil
threshold = 0.3
y_customer = (y_proba >= threshold).astype(int)
plt.hist(y_customer, bins=10)
```

```python
# learning curve
from sklearn.model_selection import learning_curve

train_sizes, train_scores, val_scores = learning_curve(
    modele, X_train, y_train,
    train_sizes=np.linspace(0.1, 1.0, 10),
    scoring='f1',
    cv=5
)

plt.plot(train_sizes, train_scores.mean(axis=1), label='train')
plt.plot(train_sizes, val_scores.mean(axis=1), label='validation')
plt.legend()
```
